```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori  
```

```python
store_data=pd.read_csv(r'store_data.csv',header=None)
store_data
```

```python
store_data.shape
```

Data Preprocessing 

```python
store_data.values
```

```python
#getting the values to a list
records=[]
for i in range(0,7501):
    records.append([str(store_data.values[i,j]) for j in range(0,20)]) 
records[1][0]
```

Applying Apriori

```python
Association_rule=apriori(records,min_support=0.0045,min_confidence=0.2,min_lift=3,min_length=2)
Association_results=list(Association_rule)
print(Association_results[0])
```

```python

for item in Association_results:
    pair= item[0]
    items=[x for x in pair]
    print(f'Rule: {items[0]}-->{items[1]}')
    #second item in list
    print(f'Support: {item[1]}')
    print(f'Confidence: {item[2][0][2]}')
    print(f'Lift: {item[2][0][3]}')
    print('========================================')
```
